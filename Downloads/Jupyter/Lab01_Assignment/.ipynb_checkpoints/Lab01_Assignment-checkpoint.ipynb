<h1>Phishing Websites: An Assessment of Features Related to Phishing Websites using an Automated Technique</h1> The Phishing Websites dataset consists of 11,055 website instances, each described by 30 numeric features aimed at identifying phishing websites. These features capture various attributes such as URL characteristics, domain registration details, presence of IP addresses, SSL states, and the use of URL-shortening services. The data is sourced primarily from public archives such as PhishTank and MillerSmiles, as well as Google search operations. Each website in the dataset is labeled to indicate whether it's a legitimate (1) site, a suspicious (0) case, or a phishing (-1) site, enabling classification tasks to detect phishing threats effectively. This dataset and its feature selection methodology are based on the work by Mohammad et al. (2012), which assessed the effectiveness of automated techniques in identifying phishing websites through a diverse set of attributes.

<h1>Bar Graph Visualization: Website Classification Counts Interpretation</h1>

The bar graph shows that the dataset includes only phishing (-1) and legitimate (1) websites. There are no suspicious (0) sites in this version of the dataset. This means it’s a binary classification problem, and the model will learn to distinguish only between phishing and legitimate sites.

```python
# Bar Graph: Class Distribution

import matplotlib.pyplot as plt
import pandas as pd
import arff

# Load the ARFF file
with open("Training Dataset.arff", 'r') as file:
    dataset = arff.load(file)

# Convert to DataFrame
df = pd.DataFrame(dataset['data'], columns=[attr[0] for attr in dataset['attributes']])

# Count the label occurrences (last column)
label_counts = df.iloc[:, -1].value_counts()

# Plotting the bar graph
plt.figure(figsize=(6, 4))
plt.bar(label_counts.index.astype(str), label_counts.values, color='skyblue', edgecolor='black')

plt.title("Website Classification Counts", fontsize=14)
plt.xlabel("Class Label (-1: Phishing, 0: Suspicious, 1: Legitimate)", fontsize=14)
plt.ylabel("Number of Websites", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()

#save the figure
plt.savefig("ChakerBaloch_BarGraphWebsite.png")

plt.show()


```

<h3>Interpretation</h3>
<p>The bar graph displays the number of websites in each classification category: phishing (-1) and legitimate (1). The dataset contains slightly more legitimate websites than phishing ones, with no entries labeled as suspicious (0).

This class distribution is fairly balanced, which is ideal for training a machine learning model, as it reduces the risk of bias toward one class. A balanced dataset helps ensure the model learns to accurately differentiate between phishing and legitimate websites rather than overfitting to the majority class.

This insight also confirms that the classification task is binary in this dataset version, focusing solely on detecting phishing versus legitimate websites.</p>

<h1>Line Graph Visualization</h1>
<p>The line graph compares how the average values of key features differ between phishing and legitimate websites. We can see that certain features show strong contrast, these differences could be valuable signals for machine learning models to detect phishing. For example, phishing sites may have higher or lower values in features related to URL structure or SSL status compared to legitimate sites.</p>


```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Split features and labels
# assume the last column is the label
feature_cols = df.columns[:-1]
label_col    = df.columns[-1]

features = df[feature_cols].apply(pd.to_numeric)  # convert only feature columns
labels   = df[label_col]                         # keep labels as strings

# 2. Recombine for grouping 
df_clean = features.copy()
df_clean[label_col] = labels

grouped = df_clean.groupby(label_col)

# 3. features to plot
selected = feature_cols[:10]   # first 10 numeric features

# 4. Compute mean per class 
means = grouped[selected].mean()

# 5. Plot
plt.figure(figsize=(10, 5))
for label in means.index:
    plt.plot(selected, means.loc[label], marker='o', label=f'Class: {label}')

plt.title("Average Feature Values by Class", fontsize=14)
plt.xlabel("Feature", fontsize=14)
plt.ylabel("Average Value", fontsize=14)
plt.legend(title="Website Class")
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("ChakerBaloch_LineGraphAvgFeatures.png")
plt.show()
```

<h3>Interpretation </h3>
<p>The graph shows that legitimate websites generally have higher averages for features like SSLfinal_State, Domain_registration_length while phishing sites score lower on these. Features like Prefix_Suffix and having_Sub_Domain show clear contrasts, making them useful for distinguishing between the two classes.</p>

<h1>Pie Chart Visualization: Website Classification Proportions</h1>
<p>This pie chart visualizes the proportion of websites classified as phishing (-1), suspicious (0), or legitimate (1). It helps us quickly see the class distribution and check for imbalance, which is especially important when training machine learning models.
</p>

```python
import matplotlib.pyplot as plt

# occurrences of each class in the Result column
label_counts = df.iloc[:, -1].value_counts().reindex([-1, 0, 1], fill_value=0)

# Labels for the pie chart
labels = ['Phishing (-1)', 'Suspicious (0)', 'Legitimate (1)']

# Colors
colors = ['#ff9999','#ffc107','#8bc34a']

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(label_counts.values, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Proportion of Website Classifications', fontsize=14)
plt.axis('equal')  # Ensures the pie chart is a circle
plt.tight_layout()
plt.savefig("ChakerBaloch_Pie_chartClassProportions.png")
plt.show()

```

<h3>Interpretation </h3>
<p>The pie chart shows that the dataset is made up of two main classes: phishing and legitimate websites. The suspicious (0) class is either absent or extremely rare in this dataset version. This highlights a binary classification task and may suggest the need for strategies like class balancing if phishing sites heavily outnumber legitimate ones.</p>

<h1>Box Chart Visualization</h1>
<p>This box chart visualizes the distribution of the web_traffic feature across website classes (phishing, suspicious, legitimate). It allows us to compare how this numeric feature varies between different types of websites and helps identify potential outliers or skewness in the data.</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert class labels to readable format for better plotting
df['Result'] = df.iloc[:, -1]
df['Result_Label'] = df['Result'].map({-1: 'Phishing', 0: 'Suspicious', 1: 'Legitimate'})

# Plot boxplot for the 'web_traffic' feature grouped by Result label
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Result_Label', y='web_traffic', palette='Set2')

plt.title("Distribution of Web Traffic by Website Classification", fontsize=14)
plt.xlabel("Website Classification", fontsize=14)
plt.ylabel("Web Traffic (Feature Value)", fontsize=14)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("ChakerBaloch_BoxPlotWebTraffic.png")
plt.show()

```

<h3>Interpretation </h3>
<p>The box plot shows that phishing websites have a wider and more varied web traffic distribution, often skewed toward lower values. In contrast, legitimate websites tend to have more consistent and higher web traffic. This suggests that web traffic can be a useful feature for distinguishing between phishing and legitimate sites.</p>
